```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)
num_rows = 10000
num_cols = 5

data = np.random.randn(num_rows, num_cols)
columns = [f'Feature_{i}' for i in range(num_cols)]

df = pd.DataFrame(data, columns=columns)

#Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Feature_0'], bins=50, alpha=0.7, color='blue')
plt.title('Histogram of Feature_0')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Feature_0'], df['Feature_1'], alpha=0.5, color='green')
plt.title('Scatter Plot of Feature_0 vs Feature_1')
plt.xlabel('Feature_0')
plt.ylabel('Feature_1')
plt.grid(True)
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
df.boxplot(column=columns)
plt.title('Box Plot of All Features')
plt.xlabel('Feature')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Correlation matrix heatmap
plt.figure(figsize=(10, 6))
correlation_matrix = df.corr()
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(num_cols), columns, rotation=45)
plt.yticks(range(num_cols), columns)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
